```python
import cv2
from google.colab.patches import cv2_imshow
from skimage import io
```

## we cannot read the image directly from url using opencv, thats why we used skimage.io

```python
# read the images
im1 = io.imread('https://raw.githubusercontent.com/fahimabrar/Machine-Learning/main/data/2d.png')
im2 = io.imread('https://raw.githubusercontent.com/fahimabrar/Machine-Learning/main/data/3d.png')
```

## insted of opencv funciton cv2.imshow we will use cv2_imshow that is imported above. Opencv imshow not worked in google colab environment, but will work fine on your local pc

```python
cv2_imshow(im1)
```

```python
cv2_imshow(im2)
```

```python
print(im1.shape)
print(im2.shape)
```

```python
# vertically concatenates images 
# of same width 
concat = cv2.hconcat([im1, im1])
cv2_imshow(concat)
```

#For caoncatinating images both images should be in the same shape
- for vertical concatination width should be same
- for horizontal concatination height should be the same

# Following block we resized the 3d (actually 2d) image and then concatinated both

```python
resized = cv2.resize(im2, (394, 262))
concat = cv2.vconcat([im1, resized])
cv2_imshow(concat)
```

```python
resized = cv2.resize(im2, (394, 262))
concat = cv2.hconcat([im1, resized])
cv2_imshow(concat)
```

# Converting color image to *Grayscale*

```python
im3 =  cv2.cvtColor(concat, cv2.COLOR_BGR2GRAY)
cv2_imshow(im3)
```

# Rotating images using imutils library

```python
import imutils
im4 = imutils.rotate(im1, angle=60)
cv2_imshow(im4)
```
