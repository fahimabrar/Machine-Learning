```python
import pandas as pd

data = pd.DataFrame(pd.read_csv("https://raw.githubusercontent.com/fahimabrar/Machine-Learning/main/data/IRIS.csv"))
```

```python
data.head()
```

```python
data.shape
```

```python
data = data.drop(columns = ['species'])
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
from scipy.cluster.hierarchy import linkage
```

# Single Linkage

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (20,10))

mergings = linkage(scaled_data, method="single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

# Complete linkage

```python
plt.figure(figsize = (20,10))
mergings = linkage(scaled_data, method="complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

# Average Linkage

```python
plt.figure(figsize = (20,10))
mergings = linkage(scaled_data, method="average", metric='euclidean')
dendrogram(mergings)
plt.show()
```

# Centroid Linkage

```python
plt.figure(figsize = (20,10))
mergings = linkage(scaled_data, method="centroid", metric='euclidean')
dendrogram(mergings)
plt.show()
```

# For details of different linkage methods, https://towardsdatascience.com/introduction-hierarchical-clustering-d3066c6b560e

```python
mergings = linkage(scaled_data, method="complete", metric='euclidean')
labels = cut_tree(mergings, n_clusters=3).reshape(-1, )
labels
```

```python
data["cluster_id"] = labels
```

```python
data
```

```python
import seaborn as sns
sns.scatterplot(data.sepal_length, data.sepal_width, hue = data.cluster_id, palette="dark")
```
